```python
import numpy as np
import pandas as pd
from tqdm import tqdm
from glob import glob
import matplotlib.pyplot as plt
import os
```

```python
from datetime import timedelta
import datetime
def round_time(dt=None, date_delta=datetime.timedelta(minutes=1), to='average'):
    """
    Round a datetime object to a multiple of a timedelta
    dt : datetime.datetime object, default now.
    dateDelta : timedelta object, we round to a multiple of this, default 1 minute.
    from:  http://stackoverflow.com/questions/3463930/how-to-round-the-minute-of-a-datetime-object-python
    """
    round_to = date_delta.total_seconds()
    if dt is None:
        dt = datetime.now()
    seconds = (dt - dt.min).seconds

    if seconds % round_to == 0 and dt.microsecond == 0:
        rounding = (seconds + round_to / 2) // round_to * round_to
    else:
        if to == 'up':
            # // is a floor division, not a comment on following line (like in javascript):
            rounding = (seconds + dt.microsecond/1000000 + round_to) // round_to * round_to
        elif to == 'down':
            rounding = seconds // round_to * round_to
        else:
            rounding = (seconds + round_to / 2) // round_to * round_to

    return dt + datetime.timedelta(0, rounding - seconds, - dt.microsecond)
```

```python
all_ecg_path = sorted(glob("/rdf/data/IMEC/minute_split/ECG/*.parquet"))
all_gsr_path = sorted(glob("/rdf/data/IMEC/minute_split/GSR/tonic/*.parquet"))
label_path = "/rdf/data/IMEC/raw/MobileQ_bn (labels)/ESM_SMI_beepnoesm_{}.xlsx"
gsr_path = "/rdf/data/IMEC/minute_split/GSR/tonic/{}.parquet"
```

```python
all_ecg_path[35]
```

```python
pesudo_min = 10
pesudo_window = timedelta(minutes=pesudo_min)
consider_timezone = False

all_ts = []
ids = []
all_segments_ecg = []
all_segments_gsr = []
all_labels = []
for path in tqdm(all_ecg_path[:36]):
    data = pd.read_parquet(path)
    uid = path.split("/")[-1].split(".")[0]
    gsr_path_individual = gsr_path.format(uid)
    label_path_individual = label_path.format(uid.lower())
    if os.path.isfile(label_path_individual) and os.path.isfile(gsr_path_individual):
        pass
    else:
        continue
        
#     ecg_segments = []
#     gsr_segments = []
#     ecg_ts = data.ts.values
#     ecg_segments = np.zeros((len(data), 7680))
#     for i, row in data.iterrows():
#         ecg_segments[i] = row.segments[::2]
#     all_segments_ecg.append(ecg_segments)
    
    gsr_individual = pd.read_parquet(gsr_path_individual)
    
    data = data.merge(gsr_individual, on='ts', suffixes=["_ecg", "_gsr"]).reset_index()
    
    label_individual = pd.read_excel(label_path_individual)
    label_individual = label_individual.dropna(subset=["mood_stress"])[["executiontime", "mood_stress"]]
    label_individual["executiontime"] = pd.to_datetime(label_individual["executiontime"]).round("T")
    label_individual.rename(columns={"executiontime":"timestamp"}, inplace=True)
    if consider_timezone:
        label_individual = label_individual.set_index("timestamp").tz_localize('Europe/Brussels').tz_convert('UTC').tz_localize(None)
        label_individual = label_individual.reset_index()
    
    individual_ecg = []
    individual_gsr = []
    for i, row in label_individual.iterrows():
        data_labeled = data[(data.ts > (row.timestamp - pesudo_window)) & (data.ts < (row.timestamp + pesudo_window))]
        if len(data_labeled) == 0:
            continue
        individual_ecg.append(np.vstack(data_labeled.segments_ecg.values)[:,::2])
        individual_gsr.append(np.vstack(data_labeled.segments_gsr.values)[:,::4])
        all_labels.extend([row.mood_stress] * len(data_labeled))
    individual_ecg = np.concatenate(individual_ecg, 0)
    ecg_thre_high = np.percentile(individual_ecg, 99)
    ecg_thre_low = np.percentile(individual_ecg, 1)
    individual_ecg[individual_ecg > ecg_thre_high] = ecg_thre_high
    individual_ecg[individual_ecg < ecg_thre_low] = ecg_thre_low
    individual_ecg = (individual_ecg - individual_ecg.min()) / (individual_ecg.max() - individual_ecg.min())
    
    individual_gsr = np.concatenate(individual_gsr, 0)

    all_segments_ecg.append(individual_ecg)
    all_segments_gsr.append(individual_gsr)
    
all_segments_ecg = np.concatenate(all_segments_ecg, 0)
all_segments_gsr = np.concatenate(all_segments_gsr, 0)
all_segments_gsr = (all_segments_gsr - all_segments_gsr.mean()) / all_segments_gsr.std()
all_segments_gsr = (all_segments_gsr - all_segments_gsr.min()) / (all_segments_gsr.max() - all_segments_gsr.min())
# ids = np.array(ids)
all_labels = np.array(all_labels)

np.save("/rdf/data/IMEC/semi_supervided/segments/window_10_min/ecg_train.npy", all_segments_ecg)
np.save("/rdf/data/IMEC/semi_supervided/segments/window_10_min/gsr_train.npy", all_segments_gsr)
np.save("/rdf/data/IMEC/semi_supervided/segments/window_10_min/label_train.npy", all_labels)
```
